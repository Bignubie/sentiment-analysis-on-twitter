```python
import random
import pandas as pd

cols = ['value', 'B', 'C', 'D', 'E', 'tweet']
data = pd.read_csv('data.csv', names=cols,usecols = range(6), encoding='iso-8859-1')
l = []
cnt = 15
for i in range(cnt):
    l.append(random.randrange(1, 1600000))

val = data["value"]
tweets = data["tweet"]

l1 = []
l2 = []

for i in l:
    if val[i]==4:
        l1.append(1);
    else:
        l1.append(0);
    l2.append(tweets[i])

dict = {'value': l1, 'tweet': l2}  

df = pd.DataFrame(dict) 
    
df.to_csv('data_of_50000.csv') 
df
```
